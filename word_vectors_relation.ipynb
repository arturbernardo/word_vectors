```python
!pip -q install gensim numpy scipy

import gensim.downloader as api
import numpy as np

# "glove-wiki-gigaword-50"
# "glove-wiki-gigaword-100"
# "glove-wiki-gigaword-200"
# "glove-wiki-gigaword-300"
# "word2vec-google-news-300"
MODEL_NAME = "word2vec-google-news-300"

print(f"Baixando modelo: {MODEL_NAME}")
wv = api.load(MODEL_NAME)

print("Dimensions:", wv.vector_size)
print("Size:", len(wv.index_to_key))

def show_related(word, topn=10):
    if word not in wv:
        print(f"'{word}' not in vocabulary.")
        return
    print(f"\nNear words '{word}':")
    for candidate, score in wv.most_similar(word, topn=topn):
        print(f"{candidate:20s} {score:.4f}")

def show_similarity(word1, word2):
    if word1 not in wv or word2 not in wv:
        print("not in vocabulary.")
        return
    score = wv.similarity(word1, word2)
    print(f"similarity('{word1}', '{word2}') = {score:.4f}")

def analogy(positive, negative=None, topn=10):
    negative = negative or []
    print(f"\nAnalogy: +{positive} -{negative}")
    for candidate, score in wv.most_similar(positive=positive, negative=negative, topn=topn):
        print(f"{candidate:20s} {score:.4f}")

def inspect_vector(word, first_n=10):
    if word not in wv:
        print(f"'{word}' not in vocabulary")
        return
    vec = wv[word]
    print(f"\nVector of '{word}' (firsts {first_n} positions):")
    print(vec[:first_n])
    print("Norm :", np.linalg.norm(vec))

# 4) Tests
show_related("life")
show_related("death")
show_similarity("life", "death")
show_similarity("life", "living")
show_similarity("king", "queen")

# Analogy
analogy(positive=["king", "woman"], negative=["man"], topn=10)

inspect_vector("life")
```

```python
analogy(positive=["princess", "old"], negative=["young"], topn=10)
analogy(positive=["princess", "young"], negative=["old"], topn=10)
analogy(positive=["misogyny", "ethnicity"], negative=["gender"], topn=10)
analogy(positive=["racism", "gender"], negative=["ethnicity"], topn=10)
```
