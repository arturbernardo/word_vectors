```python
!pip -q install gensim numpy scipy

from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

textos = [
    "o seguro de vida era barato",
    "existe vida em outros lugares do universo?",
    "a vida humana tem sentido e valor",
    "a vida biologica depende de energia",
    "o sentido da vida e uma pergunta filosofica",
    "a morte faz parte da vida",
    "uma bateria tem vida util limitada",
    "a vida util do equipamento acabou",
    "a pessoa procurava sentido para a vida",
    "o organismo depende de agua para manter a vida"
]

sentencas = [simple_preprocess(t) for t in textos]

model = Word2Vec(
    sentences=sentencas,
    vector_size=50,
    window=3,
    min_count=1,
    workers=1,
    epochs=500
)

wv = model.wv

print("\nMais similares a 'vida':")
for word, score in wv.most_similar("vida", topn=10):
    print(word, round(score, 4))

print("\nSimilaridade vida / morte:", wv.similarity("vida", "morte"))
print("Similaridade vida / util:", wv.similarity("vida", "util"))
```
